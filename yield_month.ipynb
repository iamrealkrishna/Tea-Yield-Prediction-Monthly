```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel(r"F:\misc\UPASI Project\final\finaldata.xlsx")
df.head(10)
```

```python
df.fillna(0, inplace=True)
```

```python
df.head()
```

```python
df['Yield']=df['Yield'].astype('float')
df
```

```python

df
```

```python
df['Rainfall Inches']=df['Rainfall Inches'].astype('float')
df
```

```python
df['Rainfall mm']=df['Rainfall mm'].astype('float')
df['Temperature min']=df['Temperature min'].astype('float')
df['Temperature max']=df['Temperature max'].astype('float')
df['Humidity 8:30']=df['Humidity 8:30'].astype('float')
df['Humidity 2:30']=df['Humidity 2:30'].astype('float')
df
```

```python
date = [str(y)+'-'+str(m) for y, m in zip(df.Year, df.Month)]
```

```python
df.index = pd.to_datetime(date)
```

```python
df['date'] = df.index
df['dayofyear']=df['date'].dt.dayofyear
df.drop('date',axis=1,inplace=True)
df.head()
```

```python
df['Temp'] = df[['Temperature min', 'Temperature max']].mean(axis=1)
df['Humidity'] = df[['Humidity 8:30', 'Humidity 2:30']].mean(axis=1)
df
```

```python
df
```

```python
df.plot(x='Year',y='Yield',style='.', figsize=(15,5))
```

```python
train = df.loc[df['Year'] <= 2018]
test = df.loc[df['Year'] == 2020]
df.head()
```

```python
X_train=train.drop(['Yield','Rainfall mm','Temperature min','Temperature max','Humidity 8:30','Humidity 2:30'],axis=1)
Y_train=train['Yield']
X_train
```

```python
X_test=test.drop(['Yield','Rainfall mm','Temperature min','Temperature max','Humidity 8:30','Humidity 2:30'],axis=1)
Y_test=test['Yield']
X_test
```

```python
from sklearn.svm import SVR
regressor=SVR(gamma='scale',kernel='poly',degree=10
             )
regressor.fit(X_train,Y_train)
Y_pred = regressor.predict(X_test)
```

```python
df1 = pd.DataFrame({'Actual Yield': Y_test, 'Predicted Yield': Y_pred})  
df1[df1['Predicted Yield']!=0].head(10)
```

```python
Y_test1=test['Yield']
```

```python
plt.figure(figsize=(20,10))
plt.scatter(X_test['Month'],Y_test1,color='red')
plt.plot(X_test['Month'],Y_test,color='green')
plt.plot(X_test['Month'],Y_pred,color='blue')
```

```python
df1.plot(figsize=(15,4))
df1.plot(subplots=True, figsize=(15,6))
```

```python
from sklearn.metrics import r2_score 
r2 = r2_score(Y_test, Y_pred) 
r2
```

```python
def MAPE(Y_actual,Y_Predicted):
    mape = np.mean(np.abs((Y_actual - Y_Predicted)/Y_actual))*100
    return mape
```

```python
LR_MAPE= MAPE(Y_test, Y_pred)
print("MAPE: ",LR_MAPE)
```
